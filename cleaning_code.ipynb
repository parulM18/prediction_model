```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import re
```

```python
df=pd.read_csv("flip_phones.csv")
```

```python
df
```

```python
df.drop(["page"],axis=1,inplace=True)
df['price']=df['price'].str.strip('₹')
df
```

```python
new=df['ram'].str.split("|",n=1,expand=True)
df['Ram']=new[0]
df['Rom']=new[1]
df.drop(columns=["ram"],inplace=True)
df
```

```python
new=df['Rom'].str.split("|",n=1,expand=True)
df['rom']=new[0]
df['x']=new[1]
df.drop(columns=["Rom"],inplace=True)
```

```python
new=df['product_dimention'].str.split(" ",n=1,expand=True)
df['Product_dimention in cm']=new[0]
df['x']=new[1]
df.drop(columns=["product_dimention"],inplace=True)
```

```python
new=df['camera'].str.split("|",n=1,expand=True)
df['Camera']=new[0]
df['x']=new[1]
df.drop(columns=["camera"],inplace=True)
```

```python
new=df['battery'].str.split(" ",n=1,expand=True)
df['Battery']=new[0]
df['y']=new[1]
df.drop(columns=["battery"],inplace=True)
df.drop(['y'],axis=1,inplace=True)
```

```python
new=df['Ram'].str.split(" ",n=1,expand=True)
df['Ram in GB']=new[0]
df['y']=new[1]
df.drop(columns=["Ram"],inplace=True)
```

```python
new=df['rom'].str.split(" ",n=1,expand=True)
df['Rom in GB']=new[0]
df['y']=new[1]
df.drop(columns=["rom"],inplace=True)
df.drop(['Rom in GB'],axis=1,inplace=True)
```

```python
new=df['y'].str.split(" ",n=1,expand=True)
df['Rom in GB']=new[0]
df['a']=new[1]
df.drop(columns=["y"],inplace=True)
```

```python
new=df['Camera'].str.split(" ",n=1,expand=True)
df['Back_Camera']=new[0]
df['a']=new[1]
df.drop(columns=["Camera"],inplace=True)
df['Back_Camera']=df['Back_Camera'].str.strip("MP")
```

```python
new=df['x'].str.split(" ",n=1,expand=True)
df['Front_Camera']=new[0]
df['a']=new[1]
df.drop(columns=["x"],inplace=True)
df.drop(['Front_Camera'],axis=1,inplace=True)
```

```python
new=df['a'].str.split(" ",n=1,expand=True)
df['Front_Camera']=new[0]
df['a']=new[1]
df.drop(columns=["a"],inplace=True)
```

```python
df
```

```python
df['Front_Camera']=df['Front_Camera'].str.strip("MP")
df.rename(columns={'Battery': 'Battery in mAh'}, inplace = True)
df.rename(columns={'Back_Camera': 'Back_Camera in MP'}, inplace = True)
df.rename(columns={'Front_Camera': 'Front_Camera in MP'}, inplace = True)
df
```

```python
df.to_csv("modified.csv")
```
